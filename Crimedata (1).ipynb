```python
# how many robbery cases are enrolled in hollywood area?

def count_robbery_cases_in_area(data, area_name):
    
    robbery_cases = data[(data['AREA NAME'].str.contains(area_name, case=False)) & 
                       (data['Crm Cd Desc'].str.contains('Robbery', case=False))]
    return robbery_cases.shape[0]
area = 'Hollywood'
robbery_count = count_robbery_cases_in_area(data, area)
print(f' {area} area: {robbery_count}')
```

```python
# Which areas have the highest crime rates?
def get_highest_crime_areas(data, top_n=10):
    crime_counts_by_area = data['AREA NAME'].value_counts()
    return crime_counts_by_area
highest_crime_areas = get_highest_crime_areas(data)
highest_crime_areas

```

### Visualization

```python
crime_data = pd.read_csv("C:/Users/Varty/Downloads/crime_data.csv")
def plot_most_common_weapons(crime_data, top_n=10):
    most_common_weapons = crime_data['Weapon Desc'].value_counts().head(top_n)
     # Plot the data
    plt.figure(figsize=(10, 6))
    sns.barplot(x=most_common_weapons.values, y=most_common_weapons.index)
    plt.title('Most Common Weapons Used in Crimes', fontsize=16)
    plt.xlabel('Number of Incidents', fontsize=14)
    plt.ylabel('Weapon Description', fontsize=14)
    plt.show()
plot_most_common_weapons(crime_data)
```

```python
# victim gender distribution
plt.figure(figsize=(8, 5))
gender_distribution.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Victim Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Victims')
plt.xticks(rotation=0)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# most common crimes
plt.figure(figsize=(12, 6))
most_common_crimes.plot(kind='barh', color='Red')
plt.title('Most Common Crimes')
plt.xlabel('Number of Occurrences')
plt.ylabel('Crime Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# areas with highest crime rates
plt.figure(figsize=(12, 6))
highest_crime_areas.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Areas with the Highest Crime Rates')
plt.xlabel('Area Name')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
