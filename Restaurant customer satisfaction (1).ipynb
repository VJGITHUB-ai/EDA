```python
#what is longest wait time and cuisine preferred?
data.loc[data['WaitTime'].idxmax(), 'WaitTime'] ,data.loc[data['WaitTime'].idxmax(), 'PreferredCuisine']
```

```python
#the number of males present in buisness dining and females present in casual?
def count_genders(data):
    males_business = data[(Rdata['Gender'] == 'Male') & (data['DiningOccasion'] == 'Business')].shape[0]
    females_casual = data[(Rdata['Gender'] == 'Female') & (data['DiningOccasion'] == 'Casual')].shape[0]
    return males_business, females_casual
males_business, females_casual = count_genders(data)
print(f"Number of males in business dining: {males_business}, Number of females in casual dining: {females_casual}")
```

```python
#what is the income of customer id 654 ,averagespend and waiting time?
customer_654 = data[data['CustomerID'] == 654]
income = customer_654['Income'].values[0]
average_spend = customer_654['AverageSpend'].values[0]
wait_time = customer_654['WaitTime'].values[0]

print(f"Customer ID: 654")
print(f"Income: {income}")
print(f"Average Spend: {average_spend}")
print(f"Wait Time: {wait_time} minutes")
```

```python
#which is most prefeered cuisine?
# Function to find the most preferred cuisine
def most_preferred_cuisine(data):
    return data['PreferredCuisine'].mode()[0]
# calling the function
preferred_cuisine = most_preferred_cuisine(data)
print(f"The most preferred cuisine is: {preferred_cuisine}")
```

```python
#what is the count of american cuisine?
def count_american_cuisine(data):
    return (data['PreferredCuisine'] == 'American').sum()
american_cuisine_count = count_american_cuisine(data)
print(f"The count of American cuisine is: {american_cuisine_count}")
```

```python
# Which is a cuisine got higher food rating among all ?
def highest_food_rating_cuisine(data):
    highest_rating = data['FoodRating'].max()
    cuisine = data[data['FoodRating'] == highest_rating]['PreferredCuisine'].values[0]
    return cuisine, highest_rating
cuisine, rating = highest_food_rating_cuisine(data)
print(f"The cuisine with the highest food rating is: {cuisine} with a rating of {rating}.")
```

```python
# The count of Gender age above 20
data.loc[data['Age'] > 20,'Gender'].value_counts()
```

```python
#which is the largest group  and what is the waittime and occasion?
def largest_group_details(data):
    largest_group_row = data.loc[data['GroupSize'].idxmax()]
    return largest_group_row['GroupSize'], largest_group_row['WaitTime'], largest_group_row['DiningOccasion']

# Call the function and print the result
group_size, wait_time, occasion = largest_group_details(data)
print(f"The largest group size is: {group_size}, Wait Time: {wait_time} minutes, Dining Occasion: {occasion}.")
```

### Visualization

```python
# age differneces of customers
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], bins = 25, kde=True)
plt.title('Age Differences of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a count plot for Preferred Cuisines with added color
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='PreferredCuisine', hue = 'PreferredCuisine' , order=data['PreferredCuisine'].value_counts().index, palette='viridis')  
plt.title('Count of Preferred Cuisines')
plt.xticks(rotation=90)
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.show()
```

```python
# scatter plot income vs average spend
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Income', y='AverageSpend', hue='PreferredCuisine', palette='viridis', alpha=0.7)
plt.title('Scatter Plot of Income vs Average Spend')
plt.xlabel('Income ()')
plt.ylabel('Average Spend ($)')
plt.legend(title='Preferred Cuisine', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# heatmap of income by age 
oldest_age = data['Age'].max()
print(f"The oldest age in the dataset is: {oldest_age}")
bins = [0, 20, 30, 40, 50, 60, 70, 80]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
income = data.groupby('AgeGroup')[['Income']].mean().reset_index()
heatmap_data =income.set_index('AgeGroup')
# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', cbar_kws={'label': 'income'})
plt.title('Heatmap of income by Age Group')
plt.xlabel('income')
plt.ylabel('Age Group')
plt.show()
```
