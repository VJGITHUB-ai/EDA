```python
#What is the average number of bike rentals per day?
average_rentals = data['cnt'].median()
print(average_rentals)
```

```python
#What is the highest and lowest bike rental count recorded 
highest_rentals = data['cnt'].max()
lowest_rentals = data['cnt'].min()

print(f'Highest bike rentals: {highest_rentals}')
print(f'Lowest bike rentals: {lowest_rentals}')
```

```python
#On which date did the highest number of bike rentals occur?
max_rentals = data.loc[data['cnt'].idxmax()]
max_rentals_date = max_rentals['dteday']
max_rentals_count = max_rentals['cnt']

print(f'Highest bike rentals occurred on {max_rentals_date} with a count of {max_rentals_count}.')
```

```python
highest_season = data.groupby('season')['cnt'].mean().idxmax()
average = data.groupby('season')['cnt'].mean().max()

print(f'The season with the highest average bike rental count is season {highest_season} with an average of {average:.2f} rentals.')

```

```python
#What is the average number of bike rentals in 2011 compared to 2012?
average_rentals_2011 = data[data['yr'] == 0]['cnt'].mean()
average_rentals_2012 = data[data['yr'] == 1]['cnt'].mean()

print(f'Average bike rentals in 2011: {average_rentals_2011:.2f}')
print(f'Average bike rentals in 2012: {average_rentals_2012:.2f}')
```

```python
#How many total bike rentals were recorded in each year?
total_rentals_per_year = data.groupby('yr')['cnt'].sum()
print(total_rentals_per_year)
```

```python
#What is the average count of casual users compared to registered users?
total_casual = data['casual'].sum()
total_registered = data['registered'].sum()
count_entries = len(data)

# Calculate averages
average_casual = total_casual / count_entries
average_registered = total_registered / count_entries

# Print results
print(f'Average casual users: {average_casual:.2f}')
print(f'Average registered users: {average_registered:.2f}')
```

### Visualization

```python
#to compare the average rentals across seasons.
average_rentals_per_season = data.groupby('season')['cnt'].mean()
plt.figure(figsize=(10, 6))
average_rentals_per_season.plot(kind='bar', color='skyblue')
plt.title('Average Bike Rentals Across Seasons')
plt.xlabel('Season')
plt.ylabel('Average Rentals')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

```python
#What is the relationship between humidity and bike rentals?
plt.figure(figsize=(10, 6))
plt.scatter(data['hum'], data['cnt'], alpha=0.5, color='green' )
plt.title('Scatter Plot of Humidity and Bike Rentals')
plt.xlabel('Humidity')
plt.ylabel('Number of Bike Rentals')
plt.grid(True)
plt.show()
```

```python
#daily bike rental counts using box plot
plt.figure(figsize=(8, 6))
data['cnt'].plot(kind='box', color='blue')
plt.title('Box Plot of Daily Bike Rental Counts')
plt.ylabel('Rental Count')
plt.grid(True)
plt.show()
```

```python
#casual vs. registered users in the total bike rentals 
total_casual = data['casual'].sum()
total_registered = data['registered'].sum()
# Data for the pie chart
sizes = [total_casual, total_registered]
labels = ['Casual Users', 'Registered Users']
colors = ['red', 'green']
# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Bike Rentals by User Type')
plt.axis('equal')  
plt.show()
```
